### Phase 3: Book Recommender

##### Importing librairies

```python
import time
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import imageio
import warnings
warnings.filterwarnings('ignore')
```

##### Importing dataset

```python
final_data = pd.read_csv('final_data.csv')
final_data.shape
```

```python
final_data
```

I've decided to run this code with a sample size of 20,000 due to the memory capacity of my computer. With the original dataset, it requires at least 24GB of RAM and at the moment, my computer only has 8GB and therefore I'm unable to properly run this recommender with the full dataset. 

```python
sample_size = 20000
final_data = final_data.sample(n=sample_size, replace=False, random_state=490)

final_data = final_data.reset_index()
final_data = final_data.drop('index',axis=1)
```

```python
final_data.head()
```

##### Dropping an unecessary column 

```python
final_data = final_data.drop(columns = ['Unnamed: 0'],axis=1)
final_data
```

##### Decided to put the author, title and genre in lower text for the recommender to aggregate the data to call for recommendations

```python
def clean_text(author):
    result = str(author).lower()
    return(result.replace(' ',''))
```

```python
final_data['author'] = final_data['author'].apply(clean_text)
```

```python
final_data
```

```python
final_data['title'] = final_data['title'].str.lower()
final_data
```

```python
final_data['genre'] = final_data['genre'].str.lower()
final_data
```

##### Deleted additional columns and concatenating the rest of the features together to further run the recommender and to avoid additional time lagging. 

```python
%%time
bookdata = final_data.drop(['isbn','year_published','page_count','description'],axis=1)

bookdata['data'] = bookdata[bookdata.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)

print(bookdata['data'].head())
```

Dataframe with concatenated data column

```python
bookdata
```

```python
bookdata.to_csv("bookdata.csv")
```

##### Transforming the targeted data because we're going to look to see if there is any correlation between the other book titles

```python
%%time
vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(bookdata['data'])
```

```python
%%time 
similarities = cosine_similarity(vectorized)
```

Similarities is the name of the correlation matrix between the book title variables. With the Count Vectorizer the machine meaures the likeliness between all of the words based on the number of words they have in common, using a distance measure called cosine similarity. As you can see, these values range between 0 and 1, closer to 0 being the furthest away, whereas 1 being the most similar. 

```python
print(similarities)
```

##### Creating a dataframe with the title information with its corresponding index, to further be used in the recommender

```python
bookdata= pd.read_csv('bookdata.csv')
```

```python
%%time
books = pd.DataFrame(similarities, columns=bookdata['title'], index=bookdata['title']).reset_index()

books.head(20)
```
